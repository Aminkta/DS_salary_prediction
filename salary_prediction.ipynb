## Feature Selection

More features tend to make models more complex and difficult to interpret and can often lead to models overfitting on the training data.
The ultimate objective of feature engineering is to select an optimal number of features to train and build models that generalize very well on the data and prevent overfitting.

We have applied feature engineering by creating the variable Region. After applying the "One Hot Coding" we ended-up with 1300 Variables. We will need to remove those which do not explain the dependent variable. We will use the scikit-learn library in order to detect the K best Variables 

```python
from sklearn.feature_selection import SelectKBest, f_regression
selector = SelectKBest(f_regression, k = 500)
X_new = selector.fit_transform(df_dummies.drop('Q9', axis=1), df_dummies.Q9)
names = df_dummies.drop('Q9', axis=1).columns.values[selector.get_support()]
scores = selector.scores_[selector.get_support()]
names_scores = list(zip(names, scores))
ns_df = pd.DataFrame(data = names_scores, columns=['Feat_names', 'F_Scores'])
ns_df_sorted = ns_df.sort_values(['F_Scores', 'Feat_names'], ascending = [False, True])
ns_df_sorted.iloc[0:50]
```

From the above analysis we can see the 50 variables which are more likely to be important in the model. We will keep **500** in order to build, examine and compare the Machine Learning Algorithms.  

```python
# keep the 500 most important feautures
keep=list(ns_df_sorted["Feat_names"])
keep.append("Q9")
```

```python
df_reduced = df_dummies[keep]
```

```python
df_reduced.head()
```

## Model Implementation

We will run 4 models by applying 10 folds cross validation and returning the mean and and the variance of the scores. We check the R-squared score and the RMSE. The models that we examine are **Ridge Regression, Random Forest, Gradient Boost and KNN**

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score

# split the dataset into train and test (80-20)
X_train, X_test, y_train, y_test = train_test_split( df_reduced.drop("Q9", axis=1), df_reduced.Q9, test_size=0.2, random_state=5)
```

The ridge regression's main power lies in the balancing of the bias variance tradeoff. While regular regression does minimize the sum of square error term, it leaves the variance unattended, and therefore might overfit to the training data. Ridge regression on the other hand penalizes large coefficient values which would therefore minimize variance.

```python
# Ridge Regression
from sklearn import linear_model, cross_validation
ridge = linear_model.Ridge(alpha=1, max_iter=1000)

score = cross_val_score(ridge, X_train, y_train, cv=10, scoring='r2')
scores = cross_val_score(ridge, X_train, y_train, cv=10,scoring="neg_mean_squared_error" )
rmse_scores = np.sqrt(-scores)


print("Score: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
print("RMSE: %0.2f (+/- %0.2f)" % (rmse_scores.mean(), rmse_scores.std() * 2))
```

The random forest algorithm is an ensemble learning method for classification and regression. It works by building a number of decision trees and outputs the mean prediction of those trees. Its advantages is that it handles missing values and maintains accuracy for missing data and it's less likely to overfit the model. However, it is computationally expensive and takes long to solve.

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()

score = cross_val_score(rf, X_train, y_train, cv=10, scoring='r2')
scores = cross_val_score(rf, X_train, y_train, cv=10,scoring="neg_mean_squared_error" )
rmse_scores = np.sqrt(-scores)


print("Score: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
print("RMSE: %0.2f (+/- %0.2f)" % (rmse_scores.mean(), rmse_scores.std() * 2))
```

Gradient boosting is a machine learning technique for regression and classification problems, which produces a prediction model in the form of an ensemble of weak prediction models (high bias, low variance). Boosting reduces error mainly by reducing bias.

```python
# Gradient Boost
from sklearn.ensemble import GradientBoostingRegressor

gr = GradientBoostingRegressor()

score = cross_val_score(gr, X_train, y_train, cv=10, scoring='r2')
scores = cross_val_score(gr, X_train, y_train, cv=10,scoring="neg_mean_squared_error" )
rmse_scores = np.sqrt(-scores)


print("Score: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
print("RMSE: %0.2f (+/- %0.2f)" % (rmse_scores.mean(), rmse_scores.std() * 2))
```

k-NN classification algorithm is a simple supervised learning algorithm. It uses existing cases to classify new cases. A new case is classified based on its nearest neighbors. The advantages of such a classifier is that it is easy to understand and that it adapts as new training data is collected. It, however, requires large memory for storing the entire training dataset. 

```python
# KNN 
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5)

score = cross_val_score(knn, X_train, y_train, cv=10, scoring='r2')
scores = cross_val_score(knn, X_train, y_train, cv=10,scoring="neg_mean_squared_error" )
rmse_scores = np.sqrt(-scores)


print("Score: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
print("RMSE: %0.2f (+/- %0.2f)" % (rmse_scores.mean(), rmse_scores.std() * 2))
```

# Model Selection, Testing, and Fine Tuning

The model with the highest R-Scored score and the lower RMSE was the **Gradient Boosting**. Let's see how it performs on the Test-Data Set

```python
gr.fit(X_train, y_train)
```

```python
print(f'R-squared on train dataset {gr.score(X_train, y_train)}')
print(f'R-squared on test dataset {gr.score(X_test, y_test)}')

print(f'RMSE on train dataset {np.sqrt(mean_squared_error(y_train, gr.predict(X_train)))}')
print(f'RMSE on test dataset {np.sqrt(mean_squared_error(y_test, gr.predict(X_test)))}')
```

As we can see from the Train and Test data metrics we do not have a problem of Variance (over-fitting) and neither of Bias (underfitting), since there is no difference between train and test data sets in any metric and the values are pretty good. ## Discussion About the Model

We chose the Gradient Boosting(descent) as an optimum model which belongs to the `ensemble` algorithms. This algorithm prevents the model from overfitting over the test data. It was clear in our models that the error in the train dataset was very close to the test dataset as well as to the 10 folds of the validation datasets. Also, the performance of the algorithm was better than the others and much better than that of KNN. Thus we have strong indications that there is no underfitting in the model. In our case we started with 1300 variables and around 15K observations.
In our dataset there were many discrepancies in the salaries. If we had to pick one subsample (let's say only the US participants  who are not students), this approach would have led to much lower RMSE but it would not be fair, since the Kaggle competion is refers to the whole dataset. Also we could have applied more feature engineering by merging more categorical variables, like grouping some job titles or the Age or the Educational Level and so on. Moreover we could have removed some outliers like those of `3$` annual salary or `500K$`.  
In general we built a model by detecting the most explanatory variables although there is still room of improvement.
